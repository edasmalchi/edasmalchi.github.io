```python
import geopandas as gpd
import pandas as pd
import folium
from siuba import *

import utils
```

```python
parcels = pd.read_csv('Parcels___Active_and_Retired_20241103.csv')
```

```python
ti_parcels = parcels
```

```python
ti_parcels = ti_parcels.rename(columns={'shape':'wkt'})
```

```python
ti_parcels.columns
```

```python
ti_parcels = ti_parcels >> select(_.blklot, _.mapblklot, _.date_rec_add,
                     _.date_map_add, _.project_id_add, _.police_company,
                    _.planning_district, _.zoning_code, _.wkt)
```

```python
ti_parcels = ti_parcels.dropna(subset=['wkt'])
```

```python
gdf = gpd.GeoDataFrame(data=ti_parcels, geometry=gpd.GeoSeries.from_wkt(ti_parcels.wkt))
```

```python
custom_geom = gpd.read_file('map.geojson')
```

```python
ti_ybi = custom_geom >> filter(_.name == 'ti_ybi')
```

```python
gdf = gdf.set_crs(utils.WGS84)
```

```python
gdf = gdf.to_crs(utils.CA_NAD83Albers)
ti_ybi = ti_ybi.to_crs(utils.CA_NAD83Albers)
```

```python
ti_parcels = gdf.clip(ti_ybi) >> select(-_.wkt)
```

```python
literally_water = ['1939096', '1939027', '1939092', '1939093', '1939022']
```

```python
ti_parcels = ti_parcels >> filter(-_.mapblklot.isin(literally_water))
```

```python
ti_parcels.to_file('ti_intermediate.geojson')
```

```python
ti_parcels.explore()
```
